```python
# Banyo_Sayısı sütunundaki değerlerin int türünde olup olmadığını kontrol eden bir fonksiyon
def is_int(val):
    try:
        int(val)
        return True
    except ValueError:
        return False

# Banyo_Sayısı sütununda değerlerin int olup olmadığını kontrol ederek bir maske oluşturun
mask = df_first['Banyo_Sayısı'].apply(is_int)

# Bu maskeyi kullanarak sadece değerleri int olan satırları filtreleyin
df_filtered = df_first[mask]
len(df_filtered)
```

```python
df = df_filtered.drop(['Türü', 'İlan_Numarası', 'Kategorisi','İlan_Güncelleme_Tarihi', 'Yatırıma_Uygunluk', 'İlan_Oluşturma_Tarihi', 'Takas', 'Fiyat_Durumu'], axis=1) 
# "axis=1" parametresi, sütun bazında işlem yapılmasını belirtir.
html = df.head(10).to_html()
display(HTML(html))
```

```python

# Bulunduğu Katı Dönüşüm fonksiyonu
def convert_floor_to_numeric(floor_str):
    if isinstance(floor_str, int):
        return floor_str  # Tam sayı ise doğrudan döndür
    elif 'Kat' in floor_str.lower():
        if 'Düz Giriş' in floor_str.lower() or 'Yüksek Giriş' in floor_str.lower() or 'Bahçe Katı' in floor_str.lower():
            return 0
        elif 'Kot' in floor_str.lower():
            return -int(floor_str.split()[1])
        else:
            return int(floor_str.split()[0])
    elif 'Dubleks' in floor_str.lower() or 'Müstakil' in floor_str.lower():
        return 100  # Yeni bir değer oluşturuldu
    else:
        return 0

# 'Bulunduğu_Kat', 'Binanın_Yaşı 'sütununu dönüştürme
df['Bulunduğu_Kat'] = df['Bulunduğu_Kat'].apply(convert_floor_to_numeric)

# Dönüştürülmüş veri setini kontrol etme
print(df.head())
```

Binanın yaşını 21 ve üzeriyse 22 olarak işaretledik 0(Yeni) yazıyorsa da 0 olarak değiştirilerek tüm veriler INT değeri oluşturmuş oldu


```python

# Bina Yaşı Dönüşüm fonksiyonu
def convert_building_age(age):
    age = str(age).strip()  # Önce veriyi string'e çevirip boşlukları temizle
    if age == '0 (Yeni)':  # Eğer yaş '0 (Yeni)' ise
        return 0
    elif age == '21 Ve Üzeri':  # Eğer yaş '21 Ve Üzeri' ise
        return 22
    elif age == '5-10':  # Eğer yaş '5-10' ise
        return 7
    elif age == '11-15':  # Eğer yaş '11-15' ise
        return 13
    elif age == '16-20':  # Eğer yaş '16-20' ise
        return 18
    else:
        return int(age)


# 'Binanın_Yaşı' sütununu dönüştürme
df['Binanın_Yaşı'] = df['Binanın_Yaşı'].apply(convert_building_age)

# Dönüştürülmüş veri setini kontrol etme
print(df.head())
```

### Normalizasyon Adımı

```python
# Kategorik değişkenlerin One-Hot Encoding işlemi
categorical_columns = ['Kullanım_Durumu', 'Eşya_Durumu', 'Isıtma_Tipi', 'Krediye_Uygunluk', 'Site_İçerisinde']
X_encoded = pd.get_dummies(df[categorical_columns])

# Sayısal değişkenleri ekleyin
X_encoded[['Brüt_Metrekare', 'Binanın_Yaşı', 'Binanın_Kat_Sayısı', 'Banyo_Sayısı', 'Net_Metrekare', 'Oda_Sayısı', 'Bulunduğu_Kat', 'Yaşam_endeksi', 'Nüfus']] = df[['Brüt_Metrekare', 'Binanın_Yaşı', 'Binanın_Kat_Sayısı', 'Banyo_Sayısı', 'Net_Metrekare', 'Oda_Sayısı', 'Bulunduğu_Kat', 'Yaşam_endeksi', 'Nüfus']]

# Hedef değişkeni belirleme
y = df['Fiyatı']
```

Sayısal verileri belli bir aralıkta ölçeklendirerek daha iyi performans sağladık

```python
# Min-Max scaler nesnesi oluşturma
scaler = MinMaxScaler()

# Sayısal sütunları seçme
numerical_columns = ['Brüt_Metrekare', 'Binanın_Kat_Sayısı', 'Binanın_Yaşı', 'Banyo_Sayısı', 'Net_Metrekare', 'Bulunduğu_Kat', 'Yaşam_endeksi', 'Nüfus']

# Özellikleri normalize etme
X_scaled = scaler.fit_transform(X_encoded[numerical_columns])

# Dönüştürülen veriyi DataFrame'e dönüştürme
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_columns)

# One-hot encoded sütunları koruma
X_encoded.drop(columns=numerical_columns, inplace=True)

# Normalleştirilmiş sayısal sütunları one-hot encoded sütunlarla birleştirme
X_final = pd.concat([X_encoded, X_scaled_df], axis=1)
```
