Banyo sayısı int değeri olmayanların yani sayı değil de herhangi 

```python
# Banyo_Sayısı sütunundaki değerlerin int türünde olup olmadığını kontrol eden bir fonksiyon
def is_int(val):
    try:
        int(val)
        return True
    except ValueError:
        return False

# Banyo_Sayısı sütununda değerlerin int olup olmadığını kontrol ederek bir maske oluşturun
mask = df_first['Banyo_Sayısı'].apply(is_int)

# Bu maskeyi kullanarak sadece değerleri int olan satırları filtreleyin
df_filtered = df_first[mask]
len(df_filtered)
```

### VERİ TEMİZLİĞİ YAPILDI

Veri Seti için sonuçlarının etki etmeyeceği ve bütün değerlerin aynı olduğu sütunları kaldırdım.
* Yatırıma uygunluk normalde fiyatlandırma üzerinde etkisi olur ancak veri setlerinin bir çoğunda bu değer bilinmiyor olarak geçtiği için modelin doğruluğun gerçek değerlere yakın olması adına kaldırmak daha uygun olacaktır.
* ilan tarihleri verilerin 2 ay içerisindeki değerler olduğunu ve en son aldığımızdaki değerler ne ise onun sabit olduğunu varsayarsak bizim sonuçlarımız için etkisi olmayacaktır.

```python
df = df_filtered.drop(['Türü', 'İlan_Numarası', 'Kategorisi','İlan_Güncelleme_Tarihi', 'Yatırıma_Uygunluk', 'İlan_Oluşturma_Tarihi', 'Takas', 'Fiyat_Durumu'], axis=1) 
# "axis=1" parametresi, sütun bazında işlem yapılmasını belirtir.
html = df.head(10).to_html()
display(HTML(html))
```

```python
# Bulunduğu Katı Dönüşüm fonksiyonu
def convert_floor_to_numeric(floor_str):
    if isinstance(floor_str, int):
        return floor_str  # Tam sayı ise doğrudan döndür
    elif 'Kat' in floor_str.lower():
        if 'Düz Giriş' in floor_str.lower() or 'Yüksek Giriş' in floor_str.lower() or 'Bahçe Katı' in floor_str.lower():
            return 0
        elif 'Kot' in floor_str.lower():
            return -int(floor_str.split()[1])
        else:
            return int(floor_str.split()[0])
    elif 'Dubleks' in floor_str.lower() or 'Müstakil' in floor_str.lower():
        return 100  # Yeni bir değer oluşturuldu
    else:
        return 0

# 'Bulunduğu_Kat', 'Binanın_Yaşı 'sütununu dönüştürme
df['Bulunduğu_Kat'] = df['Bulunduğu_Kat'].apply(convert_floor_to_numeric)

# Dönüştürülmüş veri setini kontrol etme
print(df.head())
```

```python
    
# Bina Yaşı Dönüşüm fonksiyonu
def convert_building_age(age):
    if age == '0(Yeni)':  # Eğer yaş '0 (Yeni)' ise
        return 0
    elif age == '21 Ve Üzeri':  # Eğer yaş '21 Ve Üzeri' ise
        return 22
    else:
        return age  # Diğer durumlarda yaş değerini olduğu gibi döndür
    
df['Binanın_Yaşı'] = df['Binanın_Yaşı'].apply(convert_building_age)
    
print(df.head())
```
